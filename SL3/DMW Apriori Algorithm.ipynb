```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = [
    ['milk','onion','nutmeg','kidney beans','eggs','yogurt'],
    ['dill','onion','nutmeg','kidney beans','eggs','yogurt'],
    ['milk','apple','kidney beans','eggs'],
    ['milk','unicorn','corn','kidney beans','yogurt'],
    ['corn','onion','onion','kidney beans','ice creams','eggs']
]
```

```python
data
```

```python
te=TransactionEncoder()
```

```python
transform_data=te.fit(data).transform(data)
```

```python
df = pd.DataFrame(transform_data, columns = te.columns_)
```

```python
df
```

```python
frequent_item = apriori(df, min_support=0.6, use_colnames=True)
```

```python
frequent_item
```

```python
result = association_rules(frequent_item, metric='confidence', min_threshold=0.7)
```
